```python
# CRIAR OBJETO
plt.figure()

# Corporation
linear=np.array([1,2,3,4,5,6,7,8])
raiz_quadrada=linear**2

# Baseline
plt.plot(linear,'-o',raiz_quadrada,'-o')
# UB
plt.plot([22,44,55],'--r')

#LEGENDA
plt.xlabel('Soma Dados')
plt.ylabel('Soma outros Dados')
plt.legend(['Baseline','Corporation','UB'])

#HACHURAR
plt.gca().fill_between(range(len(linear)),
                     linear,raiz_quadrada,
                     facecolor='blue',
                     alpha=0.25)

plt.show()
```

```python
plt.figure()

# Criar range da datas : EIXO x
datas=np.arange('2017-01-01', '2017-01-09',dtype='datetime64[D]')
# Converte pra datetime
datas=pd.to_datetime(datas)

# PLOTAGEM
plt.plot(datas,linear,'-o',datas,raiz_quadrada,'-o')

#ROTAÇÃO EIXO X
x=plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)

# LEGENDAS    
ax=plt.gca()
ax.set_xlabel('Data',fontsize=12)
ax.set_ylabel('Unidade',fontsize=12)
ax.set_title('Quadratico ($x^2$) vs Linear ($x$)',fontsize=16)
    
# IMPRIME    
plt.show()
```

# Redefine cor das barras

```python

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# PLOTAGEM
# altera a cor da barra
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')

# COR DA BARRA
# altera cor da primeira barra
bars[0].set_color('#1F77B4')

# LEGENDAS
plt.xticks(pos, languages, alpha=0.8)
plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# REMOVE VALORES EIXO Y
# Remove todos ticks (ambos eixos) e label do tick do eixo y
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)

# REMOVE MOLDURA
# retira linhas
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

## Inserir label em cada barra

```python
plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# PLOTAGEM
# altera cor da barra
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')

# COR DA BARRA
# altera cor da primeira barra
bars[0].set_color('#1F77B4')

# LEGENDAS
plt.xticks(pos, languages, alpha=0.8)
# remove the Y label since bars are directly labeled
#plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# REMOVE VALORES EIXO Y
# remove todos ticks (ambos eixos) e label do tick do eixo y
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)

# REMOVE MOLDURA
# retira linhas
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# LEGENDA BARRA    
# inserir valores em cada barra
for bar in bars:
    plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11)
plt.show()
```
