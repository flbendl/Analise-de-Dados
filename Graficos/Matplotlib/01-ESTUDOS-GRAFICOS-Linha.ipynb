```python
# Importar Bibliotecas
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[0,20,30,40,80,50,100]
mes=['Jan','Fev','Mar','Abr','Mai','Jun','Jul']
```

```python
plt.xlabel('Mês')
plt.plot(x,y,'g--');
```

```python
fig=plt.figure(figsize=(10.5,3.0))
ax=fig.add_subplot(1,1,1)

ax.set_title('Meu Primeiro Matplotlib',fontsize=16)
ax.set_xlabel('Mês',fontsize=16)
ax.set_ylabel('Vendas no Ano',fontsize=12)
ax.set_xticks(np.arange(1,8))
ax.set_xticklabels(mes,fontweight="bold",fontsize=14)

color = 'tab:red'
ax.tick_params(axis='y', labelcolor=color,width=4.5)

ax.plot(x,y,'g--',linewidth=2.0);
```

```python
# DADOS (eixo X , eixo Y)
y=[0,20,30,40,80,50,100]
mes=['Jan','Fev','Mar','Abr','Mai','Jun','Jul']
# COR (eixo Y)
cor='tab:red'
```

```python
# TAMANHO DO GRÁFICO
plt.figure(figsize=(10.5,3.0))
# PLOTAGEM 
plt.plot(mes,y,color='blue')
# TÍTULO
plt.title('Meu Primeiro Matplotlib',color='#4f5b66',fontsize=16,loc='left')

# ****** EIXO X ******
# CONFIGURAÇÃO 
plt.xlabel('Mês',fontsize=14,color='#4f5b66')
# LGENDA (string)
plt.xticks(mes,fontweight="bold",fontsize=14)

# ****** EIXO Y ******
# LEGENDA
plt.ylabel('Vendas no Ano',fontsize=14,color='#4f5b66')
# CONFIGURAÇÃO (valores)
plt.tick_params(axis='y', labelcolor=cor,length=3, width=2,labelsize='large')
               
# IMPRIME
plt.show()
```
