```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
# CARREGAR DADOS 
df=pd.read_csv('./data/gas_prices.csv')
```

```python
df.head()
```

```python
df.tail()
```

## Plotar Gráfico de Linhas

```python
# TAMANHO DO GRÁFICO
plt.figure(figsize=(10,5))

# PLOTAGEM
plt.plot(df['Year'],df['USA'],label='United States')
plt.plot(df['Year'],df['Canada'])
plt.plot(df['Year'],df['South Korea'])

# TÍTULOS
# cores : #71797b , #3987a7
plt.title('Gas Price over time in(USD)',fontdict={'fontname':'Segoe UI','fontsize':24,
                                                  'color':'#71797b'})
# LEGENDA
plt.legend(fontsize=12)

# LEGENDA DO EIXO X
plt.xticks(df['Year'][::3],fontsize=12)

plt.show()
```

## Com uso de objetos fig , ax

```python
# Criar objeto figura (fig) e ax (eixos) , tamanho da figura
fig,ax=plt.subplots(figsize=(10,5))

# PLOTAGEM
ax.plot(df['Year'],df['USA'],label='United States')
ax.plot(df['Year'],df['Canada'])
ax.plot(df['Year'],df['South Korea'])

# TÍTULO
ax.set_title('Gas Price over time in(USD)',fontsize=24)

# TÍTULO EIXO X
ax.set_xticks(df['Year'][::3])
# Tamanho da fonte do título eixo X
ax.xaxis.set_tick_params(labelsize=14)

# TÍTULO DO EIXO Y
ax.set_ylabel('UD Dollar',fontsize=16)
# Tamanho da fonte do título eixo Y
ax.yaxis.set_tick_params(labelsize=14)

# LEGENDA
ax.legend(fontsize=14)

plt.show()
```

```python
# TAMANHO DO GRÁFICO
plt.figure(figsize=(10,5))

# PLOTAGEM
plt.plot(df['Year'],df['USA'],'b.-',label='United States')
plt.plot(df['Year'],df['Canada'],'r.-')
plt.plot(df['Year'],df['South Korea'],'g.-')

# TÍTULOS
# cores : #71797b , #3987a7
plt.title('Gas Price over time in(USD)',fontdict={'fontname':'Segoe UI','fontsize':24,
                                                  'color':'#71797b'})
# LEGENDA
plt.legend(fontsize=12)

# TÍTULO DO EIXO X
plt.xlabel('Year',fontsize=16)
plt.ylabel('UD Dollar',fontsize=16)

# LEGENDA DO EIXO X
plt.xticks(df['Year'][::3],fontsize=12)

plt.show()
```

## Outra maneira de realizar plotagem

```python
# TAMANHO DO GRÁFICO
plt.figure(figsize=(10,5))

# PLOTAGEM
paises=['USA','Canada','Germany']

# Outra maneira de plotar dados no gráfico
for item in df:
    if item in paises:
        plt.plot(df['Year'],df[item])


# TÍTULOS
# cores : #71797b , #3987a7
plt.title('Gas Price over time in(USD)',fontdict={'fontname':'Segoe UI','fontsize':22,
                                                  'color':'#71797b','fontweight':'bold'})
# LEGENDA
plt.legend(fontsize=12)

# TÍTULO DO EIXO X
plt.xlabel('Year',fontsize=16)
plt.ylabel('UD Dollar',fontsize=16)

# LEGENDA DO EIXO X
plt.xticks(df['Year'][::3].tolist()+[2011],fontsize=12)

plt.show()
```

### Gráfico com Área Sombreada  (Achurado)

* Tamanho da fonte eixo y
* Altera escala do eixo y ---   `plt.ylim([escala INICIAL,escala FINAL])`

```python
# TAMANHO DO GRÁFICO
plt.figure(figsize=(10,5))

# PLOTAGEM
plt.plot(df['Year'],df['USA'],label='United States')

# TÍTULOS
# cores : #71797b , #3987a7
plt.title('Gas Price over time in(USD)',fontdict={'fontname':'Segoe UI','fontsize':24,
                                                  'color':'#71797b'})
# LEGENDA
plt.legend(fontsize=12,loc=2)

# LEGENDA DO EIXO X
plt.xticks(df['Year'][::3],fontsize=12)
# TAMANHO FONTE EIXO Y
plt.yticks(fontsize=14)
# ALTERA ESCALA DO EIXO Y
plt.ylim([0,8])

# TÍTULO DO EIXO X
plt.xlabel('Year',fontsize=16)
plt.ylabel('UD Dollar',fontsize=16)
# ÁREA SOMBREADA
plt.fill_between(df['Year'],df['USA'],alpha=0.3)

plt.show()
```
